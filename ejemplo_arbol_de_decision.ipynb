```python
# 1. Importar librerías
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 2. Crear datos de ejemplo
data = {
    'Edad': [25, 35, 45, 55, 65, 75],
    'Fuma': [0, 1, 1, 0, 1, 0],
    'Riesgo': [0, 0, 1, 1, 1, 1]
}
```

```python
df = pd.DataFrame(data)
```

```python
# 3. Separar variables predictoras (X) y objetivo (y)
X = df[['Edad', 'Fuma']]
y = df['Riesgo']
```

```python
# 4. Crear el modelo del árbol
modelo = DecisionTreeClassifier(
    criterion='entropy',  # mide la pureza (también puedes usar 'gini')
    max_depth=3,          # controla la profundidad para evitar sobreajuste
    random_state=42
)
```

```python
# 5. Entrenar el modelo
modelo.fit(X, y)
```

```python
# 6. Visualizar el árbol
plt.figure(figsize=(10,6))
plot_tree(
    modelo,
    feature_names=['Edad', 'Fuma'],
    class_names=['Bajo', 'Alto'],
    filled=True,
    rounded=True
)
plt.show()
```

```python
# 7. Hacer una predicción
nuevo_paciente = [[60, 1]]  # Edad=60, Fuma=Sí
prediccion = modelo.predict(nuevo_paciente)
print("Predicción para paciente nuevo:", "Alto riesgo" if prediccion[0]==1 else "Bajo riesgo")
```
